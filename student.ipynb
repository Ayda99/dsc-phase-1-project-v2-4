# Question 1: What is the highest gross profit obtained for movies?

```python
#What is the highest gross profit obtained for movies?
#Fill NaN values with 0 and find total gross
money["foreign_gross"]=money["foreign_gross"].fillna(0)
money["foreign_gross"] = pd.to_numeric(money["foreign_gross"], errors='coerce')
money['Total_gross'] = money['domestic_gross'] + money['foreign_gross']
money.sort_values('Total_gross', ascending= False).head(5)
```

```python
#bar graph to see highest gross profit
money.head(10).plot.bar(x="title", y="Total_gross")
plt.title("Total_gross")
plt.xlabel("")
plt.ylabel("Total_gross(billions)")
plt.show
plt.savefig("/users/aydabwg/Desktop/new_fig")
```

The total profit with domestic and foreign added together has a wide range from high earning movies in billions to low earning movies close to 100's. # Question 4: Does Genres and gross of profit have any correlation?

```python
#Create a new table to add gross pay to sql database
cursor = conn.cursor()
cursor.execute(""" CREATE TABLE profits
                (title text, 
                 studio text, 
                 domestic_gross integer,
                 foreign_gross integer,
                 total_gross integer)""")
```

```python
#add gross pay to sql database
for row in money.itertuples():
    cursor.execute('''
                INSERT INTO profits (title,studio,domestic_gross,foreign_gross,total_gross)
                VALUES (?,?,?,?,?)
                ''',
                (row.title, 
                row.studio,
                row.domestic_gross,
                row.foreign_gross,
                row.Total_gross
                ))
conn.commit()
```

```python
#Check if the table was added properly
pd.read_sql("select * from profits;",conn).head(5)
```

```python
#Does Genres and gross of profit have any correlation?
#join total_gross and genre and plot data to visualize Correlation 
pd.read_sql("""
select genres,total_gross
from movie_basics
join profits
    ON movie_basics.primary_title = profits.title
group by genres
order by total_gross DESC
;""",conn).head(15).plot.barh(x="genres", y="total_gross")
plt.title("Genre Vs Total_gross")
plt.ylabel("")
plt.xlabel("total_gross(billions)")
plt.xticks(rotation = 90)
plt.savefig("/users/aydabwg/Desktop/new_fig_3")
```

Answer: In comparing gross benefit and Genre the data indicates that romance/fantasy genre earned the most. Along the line Adventure,Drama and Documentary are in the top section compared to othe Genre.

# Question 5: What is the rating of each Genres with high gross profit?

```python
#What is the rating of each Genres with high gross profit?
pd.read_sql("""
select genres,total_gross,averagerating
from movie_basics
join profits
    ON movie_basics.primary_title = profits.title
join movie_ratings
    ON movie_basics.movie_id = movie_ratings.movie_id
group by genres
order by total_gross DESC
;""",conn).head(15).plot.scatter("genres", "total_gross",c="averagerating",cmap="bwr")
plt.title("Rating Vs Total_gross")
plt.ylabel("total_gross(billions)")
plt.xlabel("")
plt.xticks(rotation = 90)
plt.savefig("/users/aydabwg/Desktop/new_fig_4")
```

Answer: Most would expect that higher rating movies would have a higher profit gross. The data obtained for the correlation of the three variable total_gross, genre and average rating, is indicating that the highest gross profit earning genre does not necessarly have the highest average rating. This data is indicating that average rating does not have an impact on the gross profit of the movie produced.
